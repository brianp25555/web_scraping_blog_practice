## Gathering Data Practice

### Purpose

The purpose of this notebook is to practice data gathering techniques using beautiful soup. 

### Summary

This notebook takes a [webpage](https://chrisguillebeau.com/why-you-should-quit-your-job-and-travel-around-the-world/) by one of my favorite bloggers, Chris Guillebeau and using the text from that page creates a word cloud visualization of the most frequent words from that blog post.

#### Steps Involved

+ Scrape webpage html.
+ Identify webpage text.
+ Identify beginning and end of article in webpage text.
+ Create word cloud of most frequent words.

```python
# Import packages needed

# Import requests to read html source code
import requests

#Import beautifulsoup to scrape webpage
from bs4 import BeautifulSoup

#Import pandas to save information to a dataframe
import pandas as pd

# Import wordcloud and matplotlib.pyplot for word visualization

from wordcloud import WordCloud

import matplotlib.pyplot as plt
%matplotlib inline

```

```python
# Use requests library to get html source code

url ='https://chrisguillebeau.com/why-you-should-quit-your-job-and-travel-around-the-world/'
response = requests.get(url)

# Create file to store source code

with open('cb_post',mode='wb') as file:
    file.write(response.content)
    
```

```python
#Use beautifulsoup to open text of web page
# Save beginning and end of article as variables 

with open('cb_post') as file:
    soup = BeautifulSoup(file, 'lxml')
    title =str(soup.title.contents[0].split(':')[0].strip())
    body= soup.body.getText()
    start_article = body.find(title)
    end_article = body.find('###')
    article_body = body[start_article:end_article]
    

   
   
    
```

```python
# Use wordcloud to create visualization of blog text
# Use matplotlib.pyplot to display viz

wordcloud = WordCloud(max_words=60).generate(article_body)

# Create visualization for word cloud using matplotlib.pyplot
# Set interpolation to 'bilinear' to improve clarity


plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off");


```
